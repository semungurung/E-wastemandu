```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

# Load the dataset
data = pd.read_csv('//content/RobotData.csv')
```

```python
new_column_names = [
    'X_robot', 'Y_robot', 'Orientation_robot', 'Collision', 'X_candle1','Y_candle1', 'X_candle2', 'Y_candle2', 'X_candle3', 'Y_candle3', 'X_candle4', 'Y_candle4','X_speed','Y_speed'
]

data.columns = new_column_names
```

```python
import numpy as np

# Feature Engineering
# Creating distance features
data['dist_candle1'] = np.sqrt((data['X_robot'] - data['X_candle1'])**2 + (data['Y_robot'] - data['Y_candle1'])**2)
data['dist_candle2'] = np.sqrt((data['X_robot'] - data['X_candle2'])**2 + (data['Y_robot'] - data['Y_candle2'])**2)
data['dist_candle3'] = np.sqrt((data['X_robot'] - data['X_candle3'])**2 + (data['Y_robot'] - data['Y_candle3'])**2)
data['dist_candle4'] = np.sqrt((data['X_robot'] - data['X_candle4'])**2 + (data['Y_robot'] - data['Y_candle4'])**2)

# Creating angle features between robot’s orientation and each candle
data['angle_candle1'] = np.arctan2(data['Y_candle1'] - data['Y_robot'], data['X_candle1'] - data['X_robot']) - data['Orientation_robot']
data['angle_candle2'] = np.arctan2(data['Y_candle2'] - data['Y_robot'], data['X_candle2'] - data['X_robot']) - data['Orientation_robot']
data['angle_candle3'] = np.arctan2(data['Y_candle3'] - data['Y_robot'], data['X_candle3'] - data['X_robot']) - data['Orientation_robot']
data['angle_candle4'] = np.arctan2(data['Y_candle4'] - data['Y_robot'], data['X_candle4'] - data['X_robot']) - data['Orientation_robot']

# Creating relative position features
data['rel_x_candle1'] = data['X_candle1'] - data['X_robot']
data['rel_y_candle1'] = data['Y_candle1'] - data['Y_robot']
data['rel_x_candle2'] = data['X_candle2'] - data['X_robot']
data['rel_y_candle2'] = data['Y_candle2'] - data['Y_robot']
data['rel_x_candle3'] = data['X_candle3'] - data['X_robot']
data['rel_y_candle3'] = data['Y_candle3'] - data['Y_robot']
data['rel_x_candle4'] = data['X_candle4'] - data['X_robot']
data['rel_y_candle4'] = data['Y_candle4'] - data['Y_robot']

# Creating interaction features between distance and orientation
data['dist_orientation_candle1'] = data['dist_candle1'] * data['Orientation_robot']
data['dist_orientation_candle2'] = data['dist_candle2'] * data['Orientation_robot']
data['dist_orientation_candle3'] = data['dist_candle3'] * data['Orientation_robot']
data['dist_orientation_candle4'] = data['dist_candle4'] * data['Orientation_robot']

# Dropping the original candle position columns as they are now represented by new features
data.drop(columns=['X_candle1', 'Y_candle1', 'X_candle2', 'Y_candle2', 'X_candle3', 'Y_candle3', 'X_candle4', 'Y_candle4'], inplace=True)

# Checking the modified dataframe
print(data.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Assuming 'data' is your DataFrame
features = data.drop(columns=['X_speed', 'Y_speed'])  # Drop target columns from features
target_x = data['X_speed']  # Assign 'X_speed' as the target variable for X
target_y = data['Y_speed']  # Assign 'Y_speed' as the target variable for Y

# First, split the data into 80% training/validation and 20% testing
features_train_val, features_test, target_x_train_val, target_x_test, target_y_train_val, target_y_test = train_test_split(
    features, target_x, target_y, test_size=0.2, random_state=42)

# Now, split the 80% training/validation set into 70% training and 10% validation
features_train, features_val, target_x_train, target_x_val, target_y_train, target_y_val = train_test_split(
    features_train_val, target_x_train_val, target_y_train_val, test_size=0.125, random_state=42)


```

```python
# Standardize the features (important for neural networks)
scaler = StandardScaler()

# Fit the scaler on the training data and transform
features_train = scaler.fit_transform(features_train)
features_val = scaler.transform(features_val)
features_test = scaler.transform(features_test)
```

```python
from sklearn.neural_network import MLPRegressor

# Initialize the MLPRegressor model for predicting X_speed
mlp_x = MLPRegressor(
    hidden_layer_sizes=(100, 50),
     #hidden_layer_sizes=(32,16,8),
     #hidden_layer_sizes=(64,32,16),
      #hidden_layer_sizes=(256,128),
    activation='relu',
    solver='adam',
    learning_rate='adaptive',
    max_iter=200,
    random_state=42,
    batch_size=32,
    #batch_size=62,
    #batch_size=66,
)

# Initialize the MLPRegressor model for predicting Y_speed
mlp_y = MLPRegressor(
    hidden_layer_sizes=(100, 50),
     #hidden_layer_sizes=(32,16,8),
     #hidden_layer_sizes=(64,32,16),
    #hidden_layer_sizes=(256,128),
    activation='relu',
    solver='adam',
    learning_rate='adaptive',
    max_iter=200,
    random_state=42,
    batch_size=32,
    #batch_size=62,
    #batch_size=66,
)
```

```python
# Train the model for X_speed
mlp_x.fit(features_train, target_x_train)
@
# Train the model for Y_speed
mlp_y.fit(features_train, target_y_train)

```

```python

# Now, make predictions
y_pred_x = mlp_x.predict(features_test)
y_pred_y = mlp_y.predict(features_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Predict on the test set
y_pred_x = mlp_x.predict(features_test)
y_pred_y = mlp_y.predict(features_test)

# Evaluate the performance
mse_x = mean_squared_error(target_x_test, y_pred_x)
mse_y = mean_squared_error(target_y_test, y_pred_y)
r2_x = r2_score(target_x_test, y_pred_x)
r2_y = r2_score(target_y_test, y_pred_y)

print(f"X_speed - Mean Squared Error: {mse_x}, R2 Score: {r2_x}")
print(f"Y_speed - Mean Squared Error: {mse_y}, R2 Score: {r2_y}")
```
